```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import Image, display
```

### Step Activation Function
* the purpose this activation function serves is to mimic a neuron “firing” or “not firing”
* if the weights · inputs + bias​ results in a value greater than 0, the neuron will fire and output a 1; 
otherwise, it will output a 0.
* This activation function has been used historically in hidden layers, but nowadays, it is rarely a 
choice. 



```python
display(Image("images/stepfunc.png"))
```

Code formula : 

```python
def step_function(x):
    return np.where(x >= 0, 1,0)
```

```python
x = np.linspace(-10, 10, 100)
y = step_function(x)
plt.plot(x, y)
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.title("Step Function Activation")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

###  Linear Activation Function
* A linear function ( identity function ) is simply the equation of a line
* It will appear as a straight line when graphed, where y=x 
* is usually applied to the last layer’s output in the case of a __regression__ model — a model that outputs a scalar value instead of a __classification__

```python
display(Image("images/linearfunc.png"))
```

Code Formula : 

```python
def linear_activation(x):
    return x
```

```python
x = np.linspace(-10, 10, 100)
y = linear_activation(x)
plt.plot(x, y, label="f(x) = x")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.title("Linear Activation Function")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

### Sigmoid Activation Function
* The __step__ function is not ideal for neural networks because it provides minimal information (only 0 or 1), making optimization difficult

* The Sigmoid function improves this by mapping inputs smoothly between 0 and 1, retaining more information and adding nonlinearity, which is essential for neural networks. Unlike the step function, small input changes in Sigmoid lead to gradual output changes, making optimization easier

* While Sigmoid was once used in hidden layers, it was later replaced by ReLU due to efficiency. However, it remains useful in output layers, especially for binary classification tasks.

```python
display(Image("images/sigmoidfunc.png"))
```

```python
def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
y = sigmoid_function(x)

plt.plot(x, y, label=r"$\sigma(x) = \frac{1}{1+e^{-x}}$", color="blue")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.title("Sigmoid Activation Function")
plt.axhline(0, color='black', linewidth=1)  # X-axis
plt.axvline(0, color='black', linewidth=1)  # Y-axis
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

### ReLU (Rectified Linear Unit) Activation Function
* Unlike Sigmoid and Tanh, ReLU does not saturate for positive values, avoiding the vanishing gradient problem.
* It introduces nonlinearity, allowing networks to learn complex patterns.
* Computationally efficient since it involves only a simple comparison.
* Dying ReLU Problem: Neurons can get stuck outputting zero for all inputs if they receive negative gradients during training.
* A variation called Leaky ReLU addresses this by allowing a small non-zero slope for negative values.
* Sigmoid suffers from vanishing gradients in deep networks.
* ReLU speeds up training and works better in deeper architectures.

```python
display(Image("images/relu.png"))
display(Image("images/relu_.png"))
```

```python
def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
```

```python
x = np.linspace(-1, 1, 400)
print(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)

# Plot ReLU function
plt.figure(figsize=(6, 3))
plt.plot(x, y_relu, label="ReLU", linewidth=2)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("ReLU Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Output f(x)")
plt.legend()
plt.grid(True)
plt.show()

# Plot Leaky ReLU function
plt.figure(figsize=(6, 3))
plt.plot(x, y_leaky_relu, label="Leaky ReLU (alpha=0.01)", linewidth=2, color='r')
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Leaky ReLU Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Output f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

###  ReLU activation functions in hidden layers attempting to fit y=sin(x)

```python
# Generate dataset
x = np.linspace(-2 * np.pi, 2 * np.pi, 100).reshape(-1, 1)
y = np.sin(x)

# print( "x : ")
# print(x)
# print( "y : ")
# print(y)

# Initialize network parameters
np.random.seed(42)
W1 = np.random.randn(1, 8) * 0.2  # Input to first hidden layer
b1 = np.zeros((1, 8))
W2 = np.random.randn(8, 8) * -0.1  # First to second hidden layer
b2 = np.zeros((1, 8))
W3 = np.random.randn(8, 1) * 0.15  # Second hidden layer to output
b3 = np.zeros((1, 1))

# Training parameters
learning_rate = 0.01
epochs = 4000
loss_history = []

# Training loop
for epoch in range(epochs):
    # Forward pass
    h1 = relu(np.dot(x, W1) + b1)
    h2 = relu(np.dot(h1, W2) + b2)
    y_pred = np.dot(h2, W3) + b3

    # Compute loss (Mean Squared Error)
    loss = np.mean((y_pred - y) ** 2)
    loss_history.append(loss)

    # Backpropagation
    dL_dy = 2 * (y_pred - y) / y.shape[0]
    dL_dW3 = np.dot(h2.T, dL_dy)
    dL_db3 = np.sum(dL_dy, axis=0, keepdims=True)

    dL_dh2 = np.dot(dL_dy, W3.T) * relu(h2)
    dL_dW2 = np.dot(h1.T, dL_dh2)
    dL_db2 = np.sum(dL_dh2, axis=0, keepdims=True)

    dL_dh1 = np.dot(dL_dh2, W2.T) * relu(h1)
    dL_dW1 = np.dot(x.T, dL_dh1)
    dL_db1 = np.sum(dL_dh1, axis=0, keepdims=True)

    # Update weights and biases
    W3 -= learning_rate * dL_dW3
    b3 -= learning_rate * dL_db3
    W2 -= learning_rate * dL_dW2
    b2 -= learning_rate * dL_db2
    W1 -= learning_rate * dL_dW1
    b1 -= learning_rate * dL_db1

    # Display training progress
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.5f}")
        print(f"Sample Prediction at x=0: {np.dot(relu(np.dot(relu(np.dot([[0]], W1) + b1), W2) + b2), W3) + b3}")

# Display final weights and biases
print("\nFinal Weights and Biases:")
print("W1:", W1)
print("b1:", b1)
print("W2:", W2)
print("b2:", b2)
print("W3:", W3)
print("b3:", b3)

# Generate final predictions
y_pred_tanh = np.dot(relu(np.dot(relu(np.dot(x, W1) + b1), W2) + b2), W3) + b3


# Plot loss curve
plt.figure(figsize=(8, 4))
plt.plot(loss_history, label="Training Loss", color="blue")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Curve During Training")
plt.legend()
plt.grid()
plt.show()

# Plot final approximation
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="True y = sin(x)", color="blue")
plt.plot(x, y_pred_tanh, label="Trained Tanh Approximation", color="red", linestyle="dashed")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Trained Relu Network Approximation of sin(x)")
plt.legend()
plt.grid()
plt.show()
```
